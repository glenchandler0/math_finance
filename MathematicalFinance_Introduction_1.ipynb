### Where is math used in modern finance - 1?

1. Basic Algebra: Time Value of Money.
2. Basic Statistics: Understanding the relationship between prices and underlying phenomena.
3. Calculus: Risk Management; accumulation of values (integration). 
4. Linear algebra: understanding the properties of portfolios. 
5. Optimization: for portfolios. 
6. Copula functions: for modeling the joint behavior of assets with non-Gaussian correlations. 
7. Differential equations: for dynamic analysis, pricing of options. 


### Where is math used in modern finance - 2?

8. Probability and measure theory: pricing options, understanding risk preferences, correlations.  
9. Extreme value theory: for risk management, catastrophic loss. 
10. Bayesian analysis: for modeling how market makers set prices. 
11. Network theory: understanding information flows, contagion.
12. Complex Analysis: Fourier methods for pricing options.

Read Emanuel Derman's article titled "Models" in the *Financial Analysts Journal*, 2009.

### Passive vs. Active stock investing

1. Passive investing. Passive investing is often called “Buy & Hold”. You hold an index of stock (For example, an S&P 500 fund, which are the 500 largest U.S. companies) and do nothing but hold on to it no matter what happens, until you need to use the money.

2. Active investing. You or your proxy (like a mutual fund manager or financial advisor) trade in an attempt to do better than the index. There are many possible strategies in active investing.

### Active strategy number one: Time the market.

The idea here is simple: buy low and sell high. That is, buy stock when the market is low and sell stock when the market is high. Unfortunately, this is counter to human nature. Time and time again, the majority of investors who try to time the market sell low and buy high. Look how brutal the evidence for this is: 

- When the market is doing well, you want to put more into it. 
- When the market is doing poorly, you want to get out. That’s a natural urge. It’s also exactly the worst thing you can do.

### Reversion to the mean

Why do hot funds do worse than even the average fund (which, remember, does worse than the index) decade after decade after decade? This tendency for hot funds to cool back down is called reversion to the mean. We’ll talk about models for it in a later lecture.

### Why do active funds do worse than passive funds?

- Expense ratios: you pay a mutual fund manager $\approx$ 0.50–1.50% per year to trade funds vs. a fund that passively invests in an index, which costs $\approx$ 0.07–0.20% per year.
- Transaction costs: A fund loses the bid-ask spread in each trade. Worse, the bigger the mutual fund is, the more is lost in each trade, because trying to sell lots of stock brings the stock price down and trying to buy lots of stock brings the stock price up.
- Mutual funds must have some money out of the market at all times, so that they can buy desired stocks. Recall that money has a lower expected return than stocks.


### Taxes Hurt Active Funds

In taxable accounts, passive indexes do even better than indicated above because the stock trading in mutual funds triggers early capital gains taxes, as we’ll explore more next class. ### Advice for Long Term Non-professional Investors

We often get questions from students about investing money. These are important questions. Here is some further advice:

(a) As we have seen, within an asset class, there is strong evidence to stay with a passive, low cost index fund for the asset class instead of an active fund or individual stock picks.

(b) Asset allocation — that is, which asset classes you use and in what proportions — matters more than any other choice you make. It is based on your personal preferences for balancing a desire for better returns (greed) with a loathing for volatility (fear).

### Fidelity's Target Data Freedom Index Funds components

1. The U.S. Stock Index tracks the entire U.S. Stock Market.
2. The U.S. Commodities index follows commodities like oil, gold, wheat, and cotton, which can smooth out some of the volatility of stock movement and can also help to protect your portfolio against inflation. 
3. The International Stock Index follows all non-U.S. stocks. 
4. The U.S. Bond Index follows all U.S. government treasuries, corporate bonds, and mortgage backed securities. 
5. Inflation Protected Bonds, commonly called “TIPS", are government bonds that increase in worth with inflation, which gives some balance to bonds and money markets, which are vulnerable to inflation. 
6. Finally, a money market is like a bank account with slightly better returns but a little less government protection in the event of a panic.

### Rebalance Optimally

- Periodically rebalance your assets. If you want a portfolio that is 60% stocks and 40% bonds, you should keep it that way. If stocks grow so that your portfolio becomes 70% stocks after a year or more, sell stocks and buy bonds so that you have a 60-40 split again.

- In a tax-deferred or tax-free account, you should rebalance like this every year and often you can have your account set up to do rebalancing automatically. Set this up if you can. This will make you buy low and sell high automatically, taking advantage of reversion to the mean effects.
