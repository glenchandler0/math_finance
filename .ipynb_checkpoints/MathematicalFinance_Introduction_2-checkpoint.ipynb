### Efficient Markets

An Efficient Market is one that impounds all relevant information into prices accurately and instantly.

- Samuelson (1965): "Proof that Properly Anticipated Prices Fluctuate Randomly." 
- Price changes are not forecastable if all information is impounded properly. 
- Randomness is a technically rational outcome of efficiency, not a behavioral aberration. 

### Information Sets

1. Efficiency is defined with respect to an information set. 
2. You cannot make abnormal profits if all the information in the set has been impounded into prices. 

This leads to three forms of market efficiency, based on what the information set is:

- Weak form: {Past history of stock prices}. 
- Semi-strong form: {Past history of stock prices, public information}.
- Strong form: {Past history of stock prices, public information, private information}.


### Mathematics of Efficiency

1. Asset return at $t$ is $r_t$
2. For arbitrary functions $f(r_t)$ and $g(r_{t+k})$ we are interested in whether 
$$
Cov[f(r_t), g(r_{t+k})] = 0, \quad \forall k > 0
$$
3. We may further restrict $f(.), g(.)$ to be linear or arbitrary. 
4. Fair game (martingale): 
$$
E[P_{t+1} | P_t, P_{t-1}, P_{t-2}, \ldots] = P_t
$$
$$
E[P_{t+1} - P_t | P_t, P_{t-1}, P_{t-2}, \ldots] = 0
$$
i.e., non-overlapping price changes are uncorrelated at all leads and lags.
5. Note: this is not sufficient for a claim of efficiency, as compensation for risk may result in a positive drift, but if risk-adjusted, then asset prices are martingales. 

### Random Walk Hypothesis

1. This is a type of **weak-form** test.
2. Type I: iid (independent and identically distributed) increments

$$
P_t = \mu + P_{t-1} + e_t, \quad e_t \sim iid(0,\sigma^2)
$$

$$
E(P_t | P_0) = P_0 + \mu t, \quad Var(P_t | P_0) = \sigma^2 t
$$

3. Type II: Independent but not identically distributed $e_t$. 

4. Type III: uncorrelated increments, but not independent. 

$$
Cov(e_t, e_{t-k}) = 0, \quad Cov(e_t^2, e_{t-k}^2) \neq 0
$$

### Type III tests

1. Correlation and covariance? 
2. Check the autocorrelation coefficient
$$
\rho(k) = \frac{Cov(r_t,r_{t+k})}{\sqrt{Var(r_t) Var(r_{t+k})}} = 0  
$$
3. Variance Ratio Test:
$$
\mbox{is   } VR = \frac{Var(r_t + r_{t-1})}{2 Var(r_t)} = 1 
$$

### Fidelity Magellan and Peter Lynch

1. Peter Lynch beat the market in 11 years out of 13. If success is random
$$
Prob[\mbox{11 wins in 13}] = {13 \choose 11} \cdot 0.5^{11} \cdot 0.5^2 =  0.00952148
$$
That is, 1 in 105. 
2. But if Fidelity has 200 funds, then the chances that at least one fund will perform in this manner is
$$
1 - \left( \frac{104}{105} \right)^{200} = 0.852495
$$
3. And, adding in the probability of winning 12/13 or 13/13 as well, this probability increases to 0.8955. 
4. If there are 300 funds, then there is a 97% chance of one fund performing this well.

```python
#CALCULATIONS

prob = 13*12.0/2 * 0.5**11 * 0.5**2 
print(prob, 1.0/105.0)
```

```python
prob = 13.0 * 0.5**12 * 0.5**1 
print(prob)
```

```python
prob = 0.5**13 * 0.5**0 
print(prob)
```

```python
prob = 1 - (104.0/105)**200
print(prob)
```

### Mutual Fund Performance

1. We measure mutual fund performance with Jensen's $\alpha$. 
2. This is the intercept in a return regression:
$$
R_{it} = \alpha + \sum_{k=1}^n \beta_{ik} \cdot R_{kt} + e_{it}
$$
3. The $R_{k}$ are various factors that explain returns.
4. Standard factors are the equity market return, bond market return, portfolios of small stocks, etc.
